```python
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import figure
figure(num=None, figsize=(20, 20), dpi=80, facecolor='w', edgecolor='k')

img = cv2.imread('WIN_20200223_20_50_51_Pro.jpg')
#img = cv2.resize(img, (800, 500))
rows,cols,ch = img.shape
x, y = (300, 300)
pts1 = np.float32([[833,841],[1126,835], [1229,1013],[762,1017]])
pts2 = np.float32([[0+x,0+y],[400+x,0+y], [400+x, 565+y],[0+x,565+y]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(1200,1200))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
import numpy as np 
M2 = np.linalg.inv(M)
dst2 = cv2.warpPerspective(dst,M2,(1200,1200))

plt.subplot(121),plt.imshow(dst),plt.title('Input')
plt.subplot(122),plt.imshow(dst2),plt.title('Output')
plt.show()
```
